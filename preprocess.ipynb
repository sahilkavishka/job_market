```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
df = pd.read_csv("cleaned_job_market.csv")
df.head()
```

```python
df['job_title'].value_counts()
```

```python
df['experience_level'].value_counts()
```

```python
df['industry'].value_counts()
```

```python
df['city'].value_counts()
```

```python
le_job = LabelEncoder()
le_exp = LabelEncoder()
le_ind = LabelEncoder()
le_city = LabelEncoder()
```

```python
df['job_title'] = le_job.fit_transform(df['job_title'])
df['experience_level'] = le_exp.fit_transform(df['experience_level'])
df['industry'] = le_ind.fit_transform(df['industry'])
df['city'] = le_city.fit_transform(df['city'])
```

```python
df.head()
```

```python
df.to_csv("complete_preprocessing.csv", index=False)
```

```python
with open('encoders.pickle', 'wb') as file:
    pickle.dump({
        "le_job": le_job,
        "le_exp": le_exp,
        "le_ind": le_ind,
        "le_city": le_city }, file)
```
